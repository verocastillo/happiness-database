```python
# Dependencies
import pandas as pd
import os
from config import pgadmin_pw
from sqlalchemy import create_engine
from sqlalchemy.dialects import postgresql
```

```python
df = pd.read_csv("../resources/sustainability.csv")
df
```

```python
sustainability_df = df.drop(df.columns[[0,3,6,7,8,9,10,11,12,13]], axis = 1)
sustainability_df.head()
```

```python
sustainability_df = sustainability_df.rename(columns={'Country Code' : 'country_id',
                                                     sustainability_df.columns[1] : 'access_water',
                                                     sustainability_df.columns[2] : 'access_electricity',
                                                     sustainability_df.columns[3] : 'renewable_energy'})

sustainability_df.head()
```

```python
engine = create_engine(f'postgresql+psycopg2://postgres:{pgadmin_pw}@localhost:5432/sustainability_db')
connection = engine.connect()
```

```python
 engine.table_names()
```

```python
df = pd.read_csv("../resources/countries.csv")
df
```

```python
countries_df = df.drop(df.columns[[2]], axis = 1)
countries_df.head()
```

```python
countries_df = countries_df.rename(columns={'Country Code' : 'country_id',
                                            'Country' : 'country_name',
                                            })

countries_df.head()
```
